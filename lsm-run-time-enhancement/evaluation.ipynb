```python
from skimage.restoration import denoise_tv_chambolle, denoise_wavelet
from skimage.filters import median
from skimage.morphology import diamond
import matplotlib.pyplot as plt
from skimage import img_as_uint
```

```python
noisy = io.imread('present/input___PA2081b_0713_train-lsm-slow--L-12_1---94---2.tif')
target = io.imread('present/target___PA2081b_0713_train-lsm-slow--L-12_1---94---2.tif')
network = io.imread('present/network___PA2081b_0713_train-lsm-slow--L-12_1---94---2.tif')
```

```python
mf = transform.rescale(median(noisy, footprint=diamond(1)), 2, order=2)
wl = transform.rescale(denoise_wavelet(noisy, rescale_sigma=True), 2, order=2)
tv = transform.rescale(denoise_tv_chambolle(noisy, weight=0.05), 2, order=2)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 8), sharex=True, sharey=True)
ax[0, 0].imshow(transform.rescale(noisy, 2, order=0), cmap='gray')
ax[0, 0].axis('off')
ax[0, 0].set_title('Noisy')
ax[0, 1].imshow(mf, cmap='gray')
ax[0, 1].axis('off')
ax[0, 1].set_title('Median')
ax[0, 2].imshow(wl, cmap='gray')
ax[0, 2].axis('off')
ax[0, 2].set_title('Wavelet denoising')
ax[1, 0].imshow(tv, cmap='gray')
ax[1, 0].axis('off')
ax[1, 0].set_title('TV')
ax[1, 1].imshow(network, cmap='gray')
ax[1, 1].axis('off')
ax[1, 1].set_title('Network')
ax[1, 2].imshow(target, cmap='gray')
ax[1, 2].axis('off')
ax[1, 2].set_title('Clean')
```

```python
io.imsave('present/mf.tif', img_as_uint(mf))
io.imsave('present/wl.tif', img_as_uint(wl))
io.imsave('present/tv.tif', img_as_uint(tv))
```
