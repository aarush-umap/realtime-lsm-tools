```python
from glob import glob
import numpy as np
from skimage import io 
from tqdm.notebook import tqdm
import random
```

```python
img_fnames = glob('datasets/PA_2081b/clean/*.tif')
```

```python
fnames = random.sample(img_fnames, 1000)
intensity = []
for fname in tqdm(fnames):
    img = io.imread(fname)
    intensity.append(np.percentile(img, 99.8))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))
_ = plt.hist(intensity, bins=100)
```

```python
from skimage.filters import threshold_mean
threshold_mean(np.array(intensity))
```

```python
for fname in random.sample(fnames, 100):
    img = io.imread(fname)
    if np.percentile(img, 99.8) < 15000:
        plt.imshow(img, cmap='gray', vmin=0, vmax=65535)
        plt.show()
```

```python
for fname in random.sample(fnames, 100):
    img = io.imread(fname)
    if np.percentile(img, 98) > 6938:
        plt.imshow(img, cmap='gray', vmin=0, vmax=65535)
        plt.show()
```
