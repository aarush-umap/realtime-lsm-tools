```python
import autofocus_tools_loci.autofocus_tools as af
from skimage import io
import matplotlib.pyplot as plt
test_image = io.imread(r'test_data/Zstack_HistopathologySlide_CAMM_1_MMStack_crop0.ome.tif')
test_image = test_image[:,:,:,0]*.3 + test_image[:,:,:,1]*.59 + test_image[:,:,:,2]*.11 
```

```python
N, H, W = test_image.shape
```

### Threshold Absolute Gradient

```python
threshold_value = 7
z_scores = []
for z_image in test_image:
    z_scores.append(af.threshold_absolute_gradient(z_image, threshold_value))

plt.scatter([i for i in range(N)], z_scores)
```
